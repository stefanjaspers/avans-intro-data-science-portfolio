```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.close('all')
```

```python
songs = pd.read_csv('datasets/04-top-spotify-songs-10s.csv', sep=',', engine='python')
```

```python
songs.head()
```

Some more info on the columns:

- title: The song's title
- artist: The Song's artist
- top genre: The genre of the track
- year: The song's year in the Billboard
- bpm: Beats per minute, the tempo of the song
- nrgy: Energy, the energy of the song. The higher the value, the more energetic.
- dnce: Danceability. The higher the value, the easier it is to dance to this song.
- dB: Loudness. The higher the value, the louder the song
- live: Liveness. The higher the value, the more likely the song is a live recording.
- val: Valence. The higher the value, the more positive the mood of the song.
- dur: Length. The song duration.
- acous: Acousticness. The higher the value, the more acoustic the song is.
- spch: Speechiness. The higher the value, the more spoken words the song contains.
- pop: Popularity. The higher the value, the more popular the song is.

There are a lot of interesting things we can find in this data. The first thing I would like to look at is the artist. I wonder which artists have been featured the most in the top songs over the years.

# Top 10 artists

```python
import seaborn as sb
from matplotlib import rcParams
rcParams['figure.figsize'] = 12, 12
```

```python
sb.countplot(y=songs['artist'], order=songs['artist'].value_counts().iloc[:10].index)

plt.title('Top 10 most popular artists of 2010-2019')
plt.xlabel('Amount of hits')
plt.ylabel('Artist name')

plt.show()
```

Katy Perry has been featured the most: 17 times. Second place goes to Justin Bieber with a number of 16. Rihanna and Maroon 5 share third place with a total amount of 15 features in the top songs from 2010-2019.

# Top 5 genres

```python
sb.countplot(y=songs['top genre'], order=songs['top genre'].value_counts().iloc[:5].index)

plt.title('Top 5 most popular genres of 2010-2019')
plt.xlabel('Amount of songs in this list')
plt.ylabel('Genre')

plt.show()
```

You can see that pop, specifically dance pop, is the winner here, and it's not even close. Clearly an outlier.

I think it would be interesting to look at some stats over the years. For example, how have the bpm and song energy changed over the years?

# BPM and song energy over the years

```python
# take the mean bpm of all songs and group them by year.
songs_mean_bpm = songs['bpm'].groupby(songs['year']).mean()
# do the same for the song energy
songs_mean_energy = songs['nrgy'].groupby(songs['year']).mean()


# mean bpm plot
songs_mean_bpm.plot()
plt.title('Mean BPM of songs over the years')
plt.ylabel('Beats per minute')
plt.xticks(np.arange(2010, 2020, step=1))

plt.show()

# mean energy plot
songs_mean_energy.plot()
plt.title('Mean energy of songs over the years')
plt.ylabel('Energy level')
plt.xticks(np.arange(2010, 2020, step=1))

plt.show()
```

Conclusion: Both the mean bpm and song energy have gone down over the past decade. Songs have become slower and less energetic.

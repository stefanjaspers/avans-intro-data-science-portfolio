```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.close('all')
```

```python
penguins = sns.load_dataset('penguins')

penguins.head()
```

Let's look at all the correlations in our DataFrame. We will be using the Pearson correlation. The Pearson correlation measures the linear relation between two numerical variables. The result is a number between -1 and 1 where

- -1 indicates a perfect negative linear relation
- 0 indicates that there is no linear relation
- 1 indicates a perfect positive linear relation

```python
penguinsCorr = penguins.corr()
penguinsCorr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

Takeaways:

- Apart from the bill depth and bill length, all variables show some form of linear correlation, varying from 0.60 - 0.87 and -0.47 - -0.58
- There is a very strong, positive linear correlation between the body mass and flipper length

Now, we will look at scatter plots for the strongest positive and negative correlations, as well as the weakest correlation.

```python
penguins.plot(kind='scatter', x='flipper_length_mm', y='body_mass_g')

plt.title('Strongest positive correlation')
plt.xlabel('Flipper length in mm')
plt.ylabel('Body mass in grams')

plt.show()
```

We can clearly see a positive linear trend in the scatter plot: The longer the flipper length, the heavier the penguin.

```python
penguins.plot(kind='scatter', x='flipper_length_mm', y='bill_depth_mm')

plt.title('Strongest negative correlation')
plt.xlabel('Flipper length in mm')
plt.ylabel('Bill depth in mm')

plt.show()
```

Although significantly less obvious, we can still see a slight negative linear trend. The flipper length seems to go up as the bill depth decreases.

```python
penguins.plot(kind='scatter', x='bill_depth_mm', y='bill_length_mm')

plt.title('Weakest correlation')
plt.xlabel('Bill depth in mm')
plt.ylabel('Bill length in mm')

plt.show()
```

The scatters are all over the place and there is no visble correlation between the bill depth and length.
